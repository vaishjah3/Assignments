```python
def is_char_in_word(word, let):
  if let in word:
    return 1
  return 0
```

```python


def is_valid_word(word, accepted, rejected):
  idx=set(range(5))

  for k,v in accepted.items():
    idx.remove(k)
    #checks if for a particular word the index in accepted list is present, if not returns 0
    if word[k]!=v:
      return 0
    #checks if for a particular word the index in rejectedd list is present in the word
  for i in rejected:
    if i in word:
      return 0
    #checks if the letter is present only in that particular index and not other
  for i in idx:
    if word[i] in accepted.values():
      print(word[i])
      return 0
  return 1
```

```python
#for all words in the corpus, it checks which word it valid based on the accepted rejected list
#checks the total probability of all words based on accepted rejected list
def get_prob_of_evidence(accepted, rejected):
  prob=0
  for k,v in word_prob.items():
    prob=prob+v * is_valid_word(k, accepted, rejected)
  return prob
```

```python
print(get_prob_of_evidence({0:"A"}, {}))
```

```python
#for a particular character, checks if character is in the word, and if the word is valid according to accpeted rejected list
#probability that a character is valid /sum of probability of all the valid words using the accepted and rejected
def is_char_likely(char, accepted, rejected):
  prob=0
  prob_evidence= get_prob_of_evidence(accepted, rejected)
  for k,v in word_prob.items():
    numm=is_char_in_word(k, char)*  v * is_valid_word(k, accepted, rejected)
    prob=prob+ (numm/ prob_evidence)
  return prob

```

```python
def best_char(accepted, rejected):
   max_prob=0
   max_char=''
   characters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
   for i in characters:
     if i not in accepted.values():
      prob=is_char_likely(i, accepted, rejected)
      if max_prob< prob:
        max_prob=prob
        max_char=i
   return (max_prob, max_char)
```

```python
print(best_char({}, {}))
```
